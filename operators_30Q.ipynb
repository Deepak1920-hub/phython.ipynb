```python
#Reverse a three-digit number using arithmetic operators.
num = 123

last = num % 10
middle = (num // 10) % 10
first = num // 100

reverse = last * 100 + middle * 10 + first
print(reverse)
```

```python
#Find the last digit and first digit of a given number.
num = 5832

last_digit = num % 10

first_digit = num
while first_digit >= 10:
    first_digit //= 10
    print(first_digit)

print("First digit:", first_digit)
print("Last digit:", last_digit)
```

```python
#Demonstrate all assignment operators (=, +=, -=, *=, /=, //=, %= and **=) on a variable.
x = 10
print("Initial:", x)

x += 5
print("After +=:", x)

x -= 3
print("After -=:", x)

x *= 2
print("After *=:", x)

x /= 4
print("After /=:", x)

x //= 2
print("After //=:", x)

x %= 3
print("After %=:", x)

x **= 2
print("After **=:", x)
```

```python
#Count how many bits are set to 1 in a number using bitwise operations.
num = 13
count = 0
n = num

while n > 0:
    count += n & 1
    n >>= 1

print("Set bits:", count)
```

```python
#Count how many vowels in a string belong to a predefined vowel list using in
text = "education"
vowels = ["a", "e", "i", "o", "u"]
count = 0

for char in text:
    if char in vowels:
        count += 1

print("Number of vowels:", count)
```

```python
#Build a simple calculator using arithmetic and comparison operators.
a = 20
b = 5
operator = "+"

if operator == "+":
    result = a + b
elif operator == "-":
    result = a - b
elif operator == "*":
    result = a * b
elif operator == "/":
    result = a / b

print("Result:", result)

if a > b:
    print("First number is larger")
elif b > a:
    print("Second number is larger")
else:
    print("Both numbers are equal")
```

```python
#Create a grading system using comparison and logical operators.
marks = 82

if marks >= 90 and marks <= 100:
    grade = "A+"
elif marks >= 80 and marks < 90:
    grade = "A"
elif marks >= 70 and marks < 80:
    grade = "B"
elif marks >= 60 and marks < 70:
    grade = "C"
elif marks >= 0 and marks < 60:
    grade = "Fail"
else:
    grade = "Invalid marks"

print("Grade:", grade)
```
